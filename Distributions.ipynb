```python
from scipy.stats import chi2, t
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import seaborn as sns
sns.set()
import numpy as np
```

# Chi-Squared distribution

The $\chi^2$ distribution with $k\geq 1$ ($k$ being an integer) degrees of freedom is supported on $(0,\infty)$ if $k=1$ and on $[0, \infty)$ in $k>1$, and its density reads
$$
f(x) = \frac{2^{-k/2}}{\Gamma\left(\frac{k}{2}\right)}x^{k/2-1}\exp\left(-\frac{x}{2}\right).
$$

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
kk = range(1, 100, 10)
dfMid = 50
x = np.linspace(0, 100, 100)
for k in kk:
    ax.plot(x, chi2.pdf(x, k),lw=5, alpha=0.5, label='k = %.0f' %k)
plt.legend(loc='best')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10,4))
k = 100
x = np.linspace(chi2.ppf(0.01, k),chi2.ppf(0.99, k), 100)
ax.plot(x, chi2.pdf(x, k),lw=3, alpha=0.5, label='k = %.0f' %k)
plt.legend(loc='best')
plt.show()
mean, var, skew, kurt = chi2.stats(k, moments='mvsk')
print("Mean: ", mean)
print("Variance: ", var)
print("Skewness: ", skew)
print("Kurtosis: ", kurt)
```

# Student distribution

The Student distribution with $\nu>0$ degrees of freedom is supported on the whole real line, and its density reads
$$
f(x) = \frac{\Gamma\left(\frac{\nu+1}{2}\right)}{\sqrt{\pi\nu}\,\Gamma\left(\frac{\nu}{2}\right)}\left(1+\frac{x^2}{\nu}\right)^{-\frac{\nu+1}{2}}.
$$

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 4))
kk = range(1, 20, 2)
kMid = 50
x = np.linspace(t.ppf(0.001, dfMid),t.ppf(0.999, dfMid), 100)
for k in kk:
    ax.plot(x, t.pdf(x, k),lw=3, alpha=0.5, label='k = %.0f' %k)
    #df is degree of freedom
    mean, var, skew, kurt = t.stats(k, moments='mvsk')
plt.legend(loc='best')
plt.show()
```
